#### **Retention by kilometers driven per driving day**

We'll explore how the distance driven by users each day influences their likelihood to continue using the app. This would tell us if the number of kilometers driven has an influence on the user behavior.

```python
# Creating a new column 'km_per_driving_day'

df['km_per_driving_day'] = df['driven_km_drives']/df['driving_days']
```

```python
# Statistical summary of the 'km_per_driving_day' column

df['km_per_driving_day'].describe()
```

We notice that the mean value is infinity, the standard deviation is NaN, and the max value is infinity. This is because there are some zero values in the driving_days column. Pandas imputes a value of infinity in the corresponding rows of the new column because division by zero is undefined.

We now need to convert these values from infinity to zero in the 'km_per_driving_day' column. We can use np.inf to refer to a value of infinity.

```python
# Converting infinite values and replacing them with zero 

df['km_per_driving_day'] = df['km_per_driving_day'].replace(np.inf, 0)
```

```python
# Confirming that the values have been replaced

df['km_per_driving_day'].describe()
```

The maximum value is 15,420 kilometers _per drive day_. This is physically impossible. Driving 100 km/hour for 12 hours is 1,200 km. It's unlikely many people averaged more than this each day they drove, so, we will disregard rows where the distance in this column is greater than 1,200 km.

New now plot a histogram of the new 'km_per_driving_day' column, disregarding those users with values greater than 1,200 km. 

```python
# Filtering data to focus on users who drove up to 1200 kilometers and visualizing the distribution by retention status
# The histogram is color-coded to differentiate between 'retained' (yellow) and 'churned' (red) users  

df_filtered = df[df['km_per_driving_day'] <= 1200]
plt.figure(figsize=(12,7))
sns.histplot(data=df_filtered,
             x='km_per_driving_day',
             hue='label',
             multiple='fill',
             shrink=0.5,
             palette=['orange', 'red']
             )
plt.title('Distribution of Kilometers Driven Per Day by Retention Status', fontsize=18);
plt.xlabel('Kilometers Driven Per Day', fontsize=14)
plt.ylabel('Proportion of Users', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

As illustrated in the visual, there seems to be a trend where users who drive longer distances daily are more likely to churn.

### **Task 3b. Handling outliers**

Our earlier visual explorations, particularly the box plots, highlighted the presence of outliers in some variables. Notably, these outliers aren't errors but a consequence of the right-skewed nature of our data.

To ensure these outliers don't skew our subsequent analyses, it's essential to manage them effectively. One common and effective approach is to replace extreme data points with a set threshold, often determined by a specific percentile of the data.

For our dataset, I'll implement a function that identifies data points beyond the 95th percentile for a given column and replaces them with the value at this 95th percentile. This way, we preserve data integrity while minimizing the impact of extreme values. 

```python
# Function definition: Imputing outliers with 95th percentile value

def percentile_calculation(col, percent):
    percentile = df[col].quantile(percent) # calculating the percentile
    df.loc[df[col] > percentile, col] = percentile # comparing each value in the column to 0.95 percentile
    print(f'The threshold is: {round(percentile, 2):>7} | and the column name is: {col}')     
```

Next, we will apply the function to the following columns:
* `sessions`
* `drives`
* `total_sessions`
* `driven_km_drives`
* `duration_minutes_drives`

```python
# Applying the function definition to the selected columns 

columns = ['sessions','drives', 'total_sessions', 'driven_km_drives', 'duration_minutes_drives']

for col in columns:
    percentile_calculation(col, 0.95)   
```
